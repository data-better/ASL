```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a deep learning course since I entered graduate school.")
```

```python
classifier(["I took a deep learning course and it wasn't worth it."])
```

```python
classifier3 = pipeline("text-generation")
```

```python
classifier3("In this course, we will teach you how to")
```

```python
generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

```python
generator(
    "In this course, we will teach you how to",
    max_length=100,
    num_return_sequences=2,
)
```

```python
unmasker = pipeline('fill-mask')
#unmasker = pipeline('fill-mask', model='bert-base-uncased')
unmasker("Hello I'm a <mask> model.", top_k=2)
```

```python
ner = pipeline("ner", grouped_entities = True)
ner("My name is Geung-Hee Lee and I work at Korea National Open University in Seoul.")
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question = "Where do I work?",
    context="My name is Geung-Hee Lee and I work at Korea National Open University in Seoul"
)
```
