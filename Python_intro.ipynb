## Python 기초

```
# 변수형
a = 123
b = 123.456
c = "Hello World"
d = True
e = [1,2,3]
```

```
# 변수 연산
a=3;b=4
c = a+b
print(c)
d = a<b
print(d)

a="Hello";b=" World";c="!"
d=a+b+c
print(d)

a=[1,2,3];b=[4,5,6]
c=a+b
print(c)
```

```
#리스트
a=[1,2,3,4,5]
a.append(6)
print(a)
print(a[0])
print(type(a))
print(a+[7,8])
```

```
# 딕셔너리
a = {"A":1, "B":2}
print(a["A"])

a["C"] = 3
print(a)
```

```
# if문
a = 1
if a< 1:
  print("Good")
elif a<2:
  print("Better")
else:
  print("Best")
```

```
# for문
for a in [3,5, 7]:
  print(a)
print("*******")    
for a in range(3) :
  print(a)
  
print("*******")  
a=[1,2,3,4,5]  
b=[c*2 for c in a]
print(b)
```

```
# 함수

def add(a,b):
  c=a+b
  return c

print(add(1,2))
```

```
# Class

class Calc:
  def __init__(self, a):
    self.a = a
    
  def add(self,b):
    print(self.a + b)
    
  def multiply(self, b):
    print(self.a * b)
    
cc = Calc(3)
cc.add(4)
cc.multiply(4)
```

## Numpy

```
import numpy as np
a = np.array([[0,1,2],[1,2,3],[3,4,5]])

print(a)
print(np.shape(a))
print(len(a))
print(np.random.rand(5))
print('*****')
print(np.zeros((3,3)))
print(np.arange(0,1,0.1))
```

```
a = np.arange(0,12,1)
b = a.reshape(3,4)
print(b)
print('*****')
print(np.reshape(a,[3,4]))
```

```
a = np.array([[0,1,2],[1,2,3],[3,4,5]])
b = np.reshape(np.arange(0,9,1),[3,3])

print(a)
print(b)
print('***')
print(a+b) 
print(a*b) #cell 단위 곱셉
# 브로드캐스팅
c = np.array([1,2,3])
print(c)
print(a+c)
print('***')
print(a.T)   #전치행렬
print(np.sum(c)) # 합

print(np.dot(a,b)) # 행렬곱
```

##matplotlib

```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0,2*np.pi)
sin1 = np.sin(x)
cos1 = np.cos(x)

plt.plot(x,sin1, label='sin')
plt.plot(x,cos1, label='cos', linestyle="dashed")
plt.legend()

plt.show()
```

```
x  = np.random.rand(100)
y1 = np.random.rand(100)
y2 = np.random.rand(100)+1

plt.scatter(x,y1, marker="*")
plt.scatter(x,y2, marker="+")
plt.show()
```

```
img = plt.imread('tiger.jpg')
plt.imshow(img)
plt.show()
```

```
img = np.reshape(np.arange(0,25,1),[5,5])
plt.imshow(img, "gray")
plt.colorbar()
plt.show()
```

```
x = np.random.normal(0,1,1000)
plt.hist(x, bins=20)
plt.show()

print("mean : ", np.average(x))
print("std : ", np.std(x))
```
