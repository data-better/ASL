## Keras를 이용하여 logestic model 작성

* 정석으로 배우는 딥러닝 코드(p102) 참조


```
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

* Dense(input_dim=2, units=1)
     - y = $\sigma(w_1 x_1 + w_2 x_2 + b)$
     
* 확률적 경사하강법, 학습률(learning rate) lr = 0.1

```
model = Sequential()
model.add(Dense(input_dim=2, units=1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
```

* data 를 OR, AND, XOR를 마련

```
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y_OR  = np.array([[0],[1],[1],[1]])
Y_AND = np.array([[0],[0],[0],[1]])
Y_XOR = np.array([[0],[1],[1],[0]])
```

* OR 데이터 적합

```
model.fit(X,Y_OR, epochs=200, batch_size=1)
classes = model.predict(X, batch_size=1)
print('확률: ')
print(classes)
```

* AND data 분류

```
model.fit(X,Y_AND, epochs=200, batch_size=1)
```

```
classes = model.predict(X, batch_size=1)
print('확률: ')
print(classes)
```

* XOR 데이터 학습

```
model.fit(X,Y_XOR, epochs=200, batch_size=1)
```

```
classes = model.predict(X, batch_size=1)
print('확률: ')
print(classes)
```

### 은닉층 1개를 포함한 다층 퍼셉트론 모형

```
model = Sequential()

# 입력층 - 은닉층
model.add(Dense(input_dim=2, units=2))
model.add(Activation('sigmoid'))

# 은닉층 - 출력층 추가 
# model.add(Dense(input_dim=2, units=1))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
```

```
model.fit(X,Y_XOR, epochs=4000, batch_size=4)
```

```
classes = model.predict(X, batch_size=4)
print('확률: ')
print(classes)
```
